This workflow demonstrates an **advanced multi-agent research system** where:
- **Input Guardrail** validates queries before processing
- **Topic Splitter** breaks down complex queries into subtopics
- **Research Agents** work in parallel using `asyncio.gather`
- **Triage Agent** routes to synthesis or quality review
- **Optimizer Agent** evaluates if more research is needed
- **Synthesizer Agent** creates the final comprehensive report
